```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
df = pd.read_excel("social_media.xlsx")
df
```

```python
# Dataset information
df.info()
```

```python
# Descriptive statistics
df.describe()
```

```python
# Count unique values in categorical features
df['Gender'].value_counts(), df['Most_Used_Platform'].value_counts(), df['addiction_level'].value_counts()
```

```python
# Group by addiction level and compute average mental health score
df.groupby("addiction_level")["Mental_Health_Score"].mean()
```

```python
# Correlation heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Distribution of daily screen time
sns.histplot(df['daily_screen_time_min'], kde=True)
plt.title("Distribution of Daily Screen Time (min)")
plt.xlabel("Minutes")
plt.show()
```

```python
sns.boxplot(x="Gender", y="focus_score", data=df)
plt.title("Focus Score by Gender")
plt.show()
```

```python
sns.countplot(data=df, x="addiction_level", hue="Gender")
plt.title("Addiction Level by Gender")
plt.show()
```

```python
df['Most_Used_Platform'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90)
plt.ylabel("")
plt.title("Most Used Social Media Platform")
plt.show()
```

```python
platform_health = df.groupby("Most_Used_Platform")["Mental_Health_Score"].mean().sort_values()
platform_health.plot(kind="bar", color="skyblue")
plt.title("Avg Mental Health Score by Platform")
plt.ylabel("Mental Health Score")
plt.xticks(rotation=45)
plt.show()
```

```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
data =pd.read_csv("C:/Users/HP/Downloads/social_media data.csv")
print(data)
```

```python
sns.histplot(data=data, x='Age', bins=30, kde=True)
plt.title('Distribution of Likes')
plt.show()
```

```python
sns.displot(data["Country"],  kde=True, rug=True, color='r')

plt.show()
```

```python
sns.scatterplot(x="Age", y="Academic_Level", data=data, hue="Gender", style="Gender", size="Gender",palette="turbo",alpha=.7)
plt.show()
```

```python
# Histogram of social media time with KDE
sns.histplot(data['social_media_time_min'], kde=True, bins=30, color='skyblue')
plt.title("Histogram of Social Media Time (in minutes)")
plt.xlabel("Social Media Time (min)")
plt.ylabel("Frequency")
plt.show()
```

```python
# KDE Plot of focus score
sns.kdeplot(data['focus_score'], shade=True, color='orange')
plt.title("KDE Plot of Focus Score")
plt.xlabel("Focus Score")
plt.ylabel("Density")
plt.show()
```

```python
# ECDF Plot of mood score
sns.ecdfplot(data['mood_score'])
plt.title("ECDF Plot of Mood Score")
plt.xlabel("Mood Score")
plt.ylabel("Proportion â‰¤ x")
plt.show()
```

```python
# Displot of anxiety levels with KDE
sns.displot(data['anxiety_level'], kde=True, bins=25, height=4, aspect=1.5, color='mediumseagreen')
plt.title("Displot of Anxiety Level")
plt.xlabel("Anxiety Level")
plt.ylabel("Density")
plt.show()
```

```python
# Barplot: Average Daily Usage by Gender
sns.barplot(data=data, x='Gender', y='Avg_Daily_Usage_Hours', estimator='mean', palette='pastel')
plt.title("Average Daily Social Media Usage by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Daily Usage (Hours)")
plt.show()
```

```python
# Countplot: Most Used Platform
sns.countplot(data=data, x='Most_Used_Platform', palette='viridis')
plt.title("Most Used Social Media Platform")
plt.xlabel("Platform")
plt.ylabel("Number of Students")
plt.xticks(rotation=45)
plt.show()
```

```python
# Boxplot: Addicted Score by Academic Level
sns.boxplot(data=data, x='Academic_Level', y='Addicted_Score', palette='Set2')
plt.title("Addicted Score by Academic Level")
plt.xlabel("Academic Level")
plt.ylabel("Addicted Score")
plt.xticks(rotation=45)
plt.show()
```

```python
# Violinplot: Digital Wellbeing Score by Gender
sns.violinplot(data=data, x='Gender', y='digital_wellbeing_score', palette='muted')
plt.title("Digital Wellbeing Score by Gender")
plt.xlabel("Gender")
plt.ylabel("Digital Wellbeing Score")
plt.show()
```

```python
# Stripplot: Mood Score by Relationship Status
sns.stripplot(data=data, x='Relationship_Status', y='mood_score', jitter=True, palette='cool')
plt.title("Mood Score by Relationship Status")
plt.xlabel("Relationship Status")
plt.ylabel("Mood Score")
plt.xticks(rotation=15)
plt.show()
```

```python
# Catplot (Box) of Focus Score by Addiction Level
sns.catplot(data=data, x='addiction_level', y='focus_score', kind='box', palette='coolwarm')
plt.title("Focus Score by Addiction Level")
plt.xlabel("Addiction Level")
plt.ylabel("Focus Score")
plt.show()
```

```python
#sns.pairplot(df[['daily_screen_time_min', 'social_media_time_min', 'sleep_hours', 'focus_score', 'anxiety_level']], hue='Gender')
#plt.suptitle("Pairplot of Key Variables", y=1.02)
#plt.show()
# Strip column names of any leading/trailing whitespace
df.columns = df.columns.str.strip()

# Check that Gender exists now
print("Gender" in df.columns)  # Should return True

# Now plot
sns.pairplot(df[['daily_screen_time_min', 'social_media_time_min', 'sleep_hours', 'focus_score', 'anxiety_level', 'Gender']], hue='Gender')
plt.suptitle("Pairplot of Key Variables", y=1.02)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.violinplot(x='addiction_level', y='anxiety_level', data=df, palette='Set2')
plt.title("Anxiety Level by Addiction Level")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.swarmplot(x='Academic_Level', y='mood_score', data=df, hue='Gender')
plt.title("Mood Score by Academic Level and Gender")
plt.xticks(rotation=45)
plt.show()
```

```python
pivot = df.pivot_table(index='Gender', columns='Most_Used_Platform', aggfunc='size', fill_value=0)
sns.heatmap(pivot, annot=True, fmt='d', cmap='Blues')
plt.title("Count of Users by Gender and Platform")
plt.xlabel("Platform")
plt.ylabel("Gender")
plt.show()
```
